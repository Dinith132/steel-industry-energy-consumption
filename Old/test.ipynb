```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate sample data (e.g., a sine wave sequence)
timesteps = 10  # Number of time steps per sample
features = 1    # Number of features per time step
samples = 100   # Number of samples

x = np.linspace(0, 20, samples)
y = np.sin(x)

# Prepare input/output data
X_data = []
y_data = []

for i in range(len(y) - timesteps):
    X_data.append(y[i:i + timesteps])
    y_data.append(y[i + timesteps])

X_data = np.array(X_data).reshape(-1, timesteps, features)
y_data = np.array(y_data).reshape(-1, 1)

# Split into train and test sets
train_size = int(0.8 * len(X_data))
X_train, X_test = X_data[:train_size], X_data[train_size:]
y_train, y_test = y_data[:train_size], y_data[train_size:]
```
